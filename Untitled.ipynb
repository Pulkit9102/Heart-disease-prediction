```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
disease_df = pd.read_csv("framingham.csv")
disease_df.drop(columns=['education'], inplace = True, axis = 1)
disease_df.rename(columns ={'male':'Sex_male'}, inplace = True)
```

```python
disease_df.dropna(axis = 0, inplace = True)
disease_df
```

```python
X = np.asarray(disease_df[['age', 'Sex_male', 'cigsPerDay', 
                           'totChol', 'sysBP', 'glucose']])
y = np.asarray(disease_df['TenYearCHD'])

X = preprocessing.StandardScaler().fit(X).transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( 
        X, y, test_size = 0.3, random_state = 4)

print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(x='TenYearCHD', data=disease_df,
             palette="BuGn_r")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy of the model is =', 
      accuracy_score(y_test, y_pred))
```

```python
import numpy as np

# The test input (same order as your X features)
test_input = np.array([[55, 1, 15.0, 240.0, 140.0, 85.0]])

# Standardize the input using the same scaler used before
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)  # X is the original full feature set
test_input_scaled = scaler.transform(test_input)

# Use your trained model to predict (assuming it's called 'model')
prediction = logreg.predict(test_input_scaled)

print("Prediction (1 = CHD likely, 0 = unlikely):", prediction[0])
```
